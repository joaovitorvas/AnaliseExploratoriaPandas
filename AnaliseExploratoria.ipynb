# Analise Exploratoria

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_excel("/content/drive/MyDrive/Cusro_Python_Pandas_Digital_Innovation-master/datasets/AdventureWorks.xlsx")

df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Verificando receita total

round(df["Valor Venda"].sum(), 2)
```

```python
# Qual o custo total

df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python


df["Custo"].sum()
```

```python
# Vendo o lucro

df["Lucro"] = df["Valor Venda"] - df["Custo"]

df.head(1)
```

```python
# Tempo envio

df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days

df.head(1)

# Confirmando o tipo

df["Tempo Envio"].dtype
```

```python
df.dtypes
```

```python
# Verificando tempo médio por marca

df.groupby("Marca")["Tempo Envio"].mean()
```

```python
# Verificando tempo médio por marca/ano

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
# Quantidade de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Fazendo o gráfico com os dados Produto, Quantidade
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos", color="red")
plt.xlabel("Total")
plt.ylabel("Produto");
```
